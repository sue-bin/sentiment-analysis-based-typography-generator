```python
def model(target_words, n): #텍스트 인풋, 원하는 유사문장 개수를 argument로 입력

    print("...유사 문장 추출을 시작합니다...\n")
    target_words = [target_words]
     
    ### 입력된 문장에 대해
            

    ### 해당 문장을 벡터화
    vect_temp = vect.transform(target_words)

    ### 벡터화된 words에 대해서 트레인 데이터를 유사도 순으로 정렬
    sim_index = cosine_similarity(vect_temp,tfvect).argsort()[:,::-1][0]

    ### 정렬된 순서로 모집중 챌린지 데이터프레임 가져옴
    most_sim_df = data.iloc[sim_index]

    ### argument에 입력된 챌린지는 자기자신으로 추천되면 안되므로 포함될 경우 제외
    most_sim_df=most_sim_df.reset_index(drop=True)[:n]
    result = most_sim_df

                  
    print("유사 문장 추출 완료!\n")
    print("---------------------------")
    print("유사 문장 추출 결과:\n")
    for i in most_sim_df["text"]:
        print("\t"+i)
    return result
```
